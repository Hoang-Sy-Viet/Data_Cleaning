# Q2: Change the date column to the same format ‘YYYY-MM-DD’.

```python
df['join_date'].value_counts()
```

```python
def split_date(x):
    s1 = str(x).split(" ")
    if len(s1) ==2:
        s2 = s1[0] 
        return s2
    try:
        return x
    except:
        return None
```

```python
df['join_date'] = df['join_date'].apply(lambda x : split_date(x))
df['join_date'] = pd.to_datetime(df['join_date']).dt.date
```

```python
df
```

# Q3: Change the name column to the title case (e.g: Jason Mraz).

```python
df['full_name'] = df['full_name'].apply(lambda x : str(x).title())
```

```python
df
```

# Q4: Make a new “email” column with the form:{last_name}.{first_name}.{id}@yourcompany.com

```python
name_arr =[]
for i in range(0,len(df)):
    S_name = df['full_name'][i].split(" ")
    last_name = S_name[1]
    first_name = S_name[0]
    id= df['cust_id'][i]
    email = '{s1}.{s2}.{s3}@yourcompany.com'.format(s1= last_name,s2=first_name,s3=id)
    name_arr.append(email)
df['email'] = name_arr
```

```python
df
```

# Q5: Change the phone number column to the format “84……”

```python
df.phone_number.value_counts()
```

```python
df['phone_number'] = df['phone_number'].astype(str)
```
